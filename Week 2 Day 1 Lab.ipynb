### Problem 2
Consider the function from Friday's quiz:
```python
def squares(maximum):

    container = []

    for i in range(maximum):
        container.append(i)

    container2 = []

    for j in container:
        container2.append(j**2)

    container3 = 0

    for k in container2:
        container3 += k

    return container3
```

What is the complexity? Justify your answer. Write a new, simpler function that does the same as `squares()`. Calculate your function's complexity and justify your answer.

````python
#The complexity of this code would be O(n) + O(n) + O(n) = O(n)
def squares(max):
    total = 0
    for i in range(max):
        total += i**2
    return total

#The complexity of this. function would be O(n) because only a single loop is used.
````

### Problem 5
Write a recursive function `fibonacci(n)` that computes the $n$-th Fibonacci number. Compute the time complexity of your `fibonacci(n)` function.

````python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
````

### Problem 6
Write a recursive function `reverse(string)` that reverses the order of a string. Compute the time complexity of your `reverse()` function

````python
def reverse(string):
    if len(string) <= 1:
        return string
    return reverse(string[1:]) + string[0]
````

### Problem 7
Write a recursive function that computes the sum of integers up to input $n$.

````python
def sum_integers(n):
    if n == 0:
        return 0
    return n + sum_integers(n - 1)
````
