### Problem 1
1. Create your own dataframe with at least 3 columns and 4 rows.
2. Save the dataframe to a csv file.
3. Read the csv back in.
4. Plot the data.

````python
import pandas as pd
import matplotlib.pyplot as plt

students = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "Grade": ["A", "B", "C", "B", "A"]
}
df = pd.DataFrame(students)
df.to_csv("students.csv", index=False)
new_df = pd.read_csv("students.csv")
print(new_df)

new_df.plot(kind='bar', x='Name', y='Age', title='Student Ages')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
````
